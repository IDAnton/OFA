```python
import numpy as np  
from math import *
import matplotlib.pyplot as plt

L = 15
n = 1000
k = 10

h = L / n

l = []

A = np.zeros((n + 1, n + 1))

t = np.linspace(0, k, k)
x = np.linspace(-0.5 * L, 0.5 * L, n + 1)
v = np.zeros((k + 1, n + 1))
u = np.zeros(n + 1)
solve = np.zeros(n + 1)

for j in range(1, n):
    A[j][j - 1] = -1 / (2 * h**2)
    A[j][j] = 1/ h**2 + 0.5 * x[j]**2
    A[j][j + 1] = -1 / (2 * h**2)

A[0][0] = 1/ h**2 + 0.5 * x[0]**2
A[0][1] = -1 / (2 * h**2)

A[n][n - 1] = -1 / (2 * h**2)
A[n][n] = 1/ h**2 + 0.5 * x[n]**2

A_inv = np.linalg.inv(A)

for j in range(0, n + 1):
    solve[j] = exp( -0.5 * x[j]**2)

solve = solve / np.linalg.norm(solve)

for j in range(0, n + 1):
    v[0][j] =  (j - n / 2)**2

for m in range(1, k + 1):
    v[m] = A_inv.dot(v[m - 1])

    l.append(np.linalg.norm(v[m - 1]) / np.linalg.norm(v[m]))

plt.plot(t, l, ':b')
plt.title('l(t)')
plt.show()

u = v[k] / np.linalg.norm(v[k])

plt.plot(x, u, ':b', label = 'u')
plt.plot(x, solve, ':g', label = 'solve')
plt.legend()
plt.show()
```
